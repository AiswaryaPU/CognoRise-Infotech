**Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Dataset Loading**

```python
df = pd.read_csv('/content/cereal.csv')
```

**Data Cleaning**

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Check for any missing values in the dataset
df.isnull().sum()
```

```python
df.columns
```

**Data Visualization**

```python
# Plot histograms for numerical columns
numerical_cols = ['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins', 'shelf', 'weight', 'cups', 'rating']
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)
    plt.xlabel(col.capitalize())
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {col.capitalize()}')
    plt.grid(True)
    plt.show()
```

```python
# Count the number of cereals from each manufacturer
manufacturer_counts = df['mfr'].value_counts()

# Create a pie chart with different colors
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink', 'lightyellow', 'lightgrey']

plt.figure(figsize=(8, 6))
plt.pie(manufacturer_counts, labels=manufacturer_counts.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Cereals by Manufacturer')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Descriptive Statistics**

```python
# Compute mean, median, and standard deviation of 'calories' column
mean_calories = df['calories'].mean()
median_calories = df['calories'].median()
std_calories = df['calories'].std()

print("Mean Calories:", mean_calories)
print("Median Calories:", median_calories)
print("Standard Deviation Calories:", std_calories)
```

**Grouping and Aggregation**

```python
# Group by 'type' and calculate mean values for numerical columns
type_mean = df.groupby('type')[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins', 'shelf', 'weight', 'cups', 'rating']].mean()

print("Mean Values by Cereal Type:")
print(type_mean)
```

```python
# Group by 'mfr' and calculate sum of 'protein' and 'fiber'
mfr_sum = df.groupby('mfr')[['protein', 'fiber']].sum()

print("\nSum of Protein and Fiber by Manufacturer:")
print(mfr_sum)
```

**Sorting and Filtering**

```python
# Sort cereals by 'rating' in descending order
sorted_by_rating = df.sort_values(by='rating', ascending=False)

print("Cereals Sorted by Rating:")
print(sorted_by_rating.head())
```

```python
# Sort cereals by 'calories' in ascending order
sorted_by_calories = df.sort_values(by='calories')

print("\nCereals Sorted by Calories:")
print(sorted_by_calories.head())
```

```python
# Filter cereals with calories less than or equal to 100 and sort by rating
low_calorie_high_rating = df[df['calories'] <= 100].sort_values(by='rating', ascending=False)

# Display the filtered and sorted dataset
print("Low Calorie Cereals with High Ratings:")
print(low_calorie_high_rating.head())
```

**Calories vs Cereal**

```python
# Box plot for comparing 'calories' by 'type'
plt.figure(figsize=(8, 6))
sns.boxplot(x='type', y='calories', data=df)
plt.xlabel('Cereal Type')
plt.ylabel('Calories')
plt.title('Comparison of Calories by Cereal Type')
plt.grid(True)
plt.show()
```

**Vitamins vs Rating**

```python
# Scatter plot for 'vitamins' vs 'rating'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='vitamins', y='rating', data=df)
plt.xlabel('Vitamins Percentage')
plt.ylabel('Rating')
plt.title('Scatter Plot: Vitamins vs Rating')
plt.grid()
plt.show()
```
